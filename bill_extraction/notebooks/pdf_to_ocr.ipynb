```python
import cv2, json, os
from pdf2image import convert_from_path
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

poppler_path = r"C:\poppler-23.11.0\Library\bin"   # <-- CHANGE to your poppler bin path
```

```python
pdf_path = "C:/Users/Vidhi Jain/Desktop/bajaj/bill_extraction/data/raw/Sample Document 1.pdf"

pages = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)

os.makedirs("C:/Users/Vidhi Jain/Desktop/bajaj/bill_extraction/data/images", exist_ok=True)

for i, p in enumerate(pages):
    img_path = f"C:/Users/Vidhi Jain/Desktop/bajaj/bill_extraction/data/images/page_{i+1}.png"
    p.save(img_path, "PNG")
    print("Saved:", img_path)
```

```python
img = cv2.imread("C:/Users/Vidhi Jain/Desktop/bajaj/bill_extraction/data/images/page_1.png")

ocr_data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

# Preview first 30 words
for i in range(30):
    print(ocr_data['text'][i], "| conf:", ocr_data['conf'][i])
```
